### What is Tokenization?

Tokenization is one of the fundamental steps in Natural Language Processing (NLP). It's the process of breaking down a text into smaller units called "tokens." These tokens can be words, subwords, characters, or even punctuation marks.

**Why is Tokenization Important?**

1.  **Meaningful Units**: Most NLP tasks (like sentiment analysis, machine translation, text classification) operate on individual words or phrases, not raw continuous text. Tokenization converts raw text into these meaningful units.
2.  **Vocabulary Creation**: It helps in building a vocabulary of unique words present in the corpus, which is essential for many statistical and machine learning models.
3.  **Feature Extraction**: Tokens serve as features for various NLP models. For example, in a Bag of Words model, the frequency of tokens is counted.
4.  **Language Understanding**: It helps in understanding the grammatical structure and semantic meaning of sentences by separating words and punctuation.

For example, the sentence "Hello, world!" might be tokenized into `['Hello', ',', 'world', '!']`.

## Text Preprocessing: Stemming and Lemmatization

Beyond basic tokenization, normalizing words to their base forms is crucial for many NLP tasks. This helps in treating words like 'run', 'running', 'runs', and 'runner' as variations of the same core concept, thereby reducing the vocabulary size and improving model performance. Two common techniques for this are **Stemming** and **Lemmatization**.

### What is Stemming?

Stemming is a basic text normalization technique that reduces words to their root or base form, known as a "stem." The stem itself may not be a grammatically correct word; its main purpose is to group together words with similar meanings. Stemming algorithms often work by chopping off suffixes from words.

**Examples of Stemming (using Porter Stemmer):**
*   `running` → `run`
*   `flies` → `fli`
*   `universal` → `univers`
*   `studies` → `studi`

### What is Lemmatization?

Lemmatization, unlike stemming, is a more sophisticated process that reduces words to their base or dictionary form, known as a "lemma." The lemma is always a grammatically correct word. Lemmatization uses vocabulary and a morphological analysis of words, often considering the word's part of speech, to achieve this.

**Examples of Lemmatization (using WordNetLemmatizer):**
*   `running` → `run` (verb)
*   `ran` → `run` (verb)
*   `better` → `good` (adjective)
*   `geese` → `goose` (noun)

### Stemming vs. Lemmatization: Key Differences

| Feature           | Stemming                                   | Lemmatization                                      |
| :---------------- | :----------------------------------------- | :------------------------------------------------- |
| **Output**        | A "stem" (may not be a real word)        | A "lemma" (always a real word)                   |
| **Approach**      | Heuristic rules, chops off suffixes        | Dictionary lookup, morphological analysis          |
| **Accuracy**      | Faster, but less accurate                  | Slower, but more accurate (linguistically informed)|
| **Use Case**      | Information retrieval, quick analysis      | Text classification, machine translation, chatbots |

In essence, if linguistic accuracy is paramount and computational resources allow, lemmatization is generally preferred over stemming.

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet

# Download necessary NLTK data for lemmatization if not already downloaded
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)

# Initialize stemmer and lemmatizer
porter_stemmer = PorterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()

# Define example words
example_words = [
    "running", "runs", "ran", "runner",
    "better", "best", "good",
    "geese", "goose",
    "studies", "studying", "study",
    "corpora", "corpus"
]

print("--- Stemming and Lemmatization on Example Words ---")
for word in example_words:
    stem = porter_stemmer.stem(word)
    # For lemmatization, we need to provide a part-of-speech (POS) tag for better accuracy
    # default is 'n' (noun), 'v' for verb, 'a' for adjective, 'r' for adverb
    # We'll try to infer POS or use a default for demonstration
    lemma = wordnet_lemmatizer.lemmatize(word, pos=wordnet.VERB) # Try as verb first
    if lemma == word: # If lemmatizing as verb didn't change it, try as noun
        lemma = wordnet_lemmatizer.lemmatize(word, pos=wordnet.NOUN)
    if lemma == word: # If still no change, try as adjective
        lemma = wordnet_lemmatizer.lemmatize(word, pos=wordnet.ADJ)

    print(f"Original: {word:<10} | Stem: {stem:<10} | Lemma: {lemma:<10}")

print("\nRequired NLTK resources for stemming and lemmatization initialized.")
```

```python
print("\n--- Stemming and Lemmatization on a Sentence ---")

# Reuse the previously tokenized sentence
# example_sentence = df["text"].iloc[0]
# tokenized_sentence = word_tokenize(example_sentence) # Assuming this was already executed and 'tokenized_sentence' is available

print(f"Original Tokenized Sentence:\n{tokenized_sentence}")

# Apply stemming to the tokenized sentence
stemmed_sentence = [porter_stemmer.stem(word) for word in tokenized_sentence]
print(f"\nStemmed Sentence:\n{stemmed_sentence}")

# Apply lemmatization to the tokenized sentence
# For better lemmatization, we'll try to determine the part-of-speech (POS) tag for each word.
# NLTK's `pos_tag` is needed for this.
import nltk
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
from nltk.tag import pos_tag

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN # Default to noun if POS is not found or ambiguous

lemmatized_sentence = []
for word, tag in pos_tag(tokenized_sentence):
    wntag = get_wordnet_pos(tag)
    lemmatized_sentence.append(wordnet_lemmatizer.lemmatize(word, wntag))

print(f"\nLemmatized Sentence:\n{lemmatized_sentence}")
```

## Feature Extraction: Bag of Words (BoW)

In Natural Language Processing, text data needs to be converted into numerical representations before machine learning algorithms can process it. One of the simplest and most fundamental ways to achieve this is through the **Bag of Words (BoW)** model.

### What is Bag of Words (BoW)?

The Bag of Words model is a representation that simplifies text by discarding grammar and even word order, but keeping the multiplicity of words. It treats each document (e.g., sentence, paragraph, or entire text) as an unordered collection (a "bag") of words.

**How it works:**

1.  **Vocabulary Creation**: First, it creates a vocabulary of all unique words found across all the documents in the corpus.
2.  **Vector Representation**: For each document, it then counts the occurrences of each word from the vocabulary. This count becomes the value for that word's position in a numerical vector representing the document.

For example, consider two simple sentences:
*   Document 1: "I love natural language processing."
*   Document 2: "Natural language processing is fun."

The vocabulary would be: `['I', 'love', 'natural', 'language', 'processing', 'is', 'fun']`.

Their BoW vectors might look like this:
*   Document 1: `[1, 1, 1, 1, 1, 0, 0]` (counts for 'I', 'love', 'natural', 'language', 'processing', 'is', 'fun')
*   Document 2: `[0, 0, 1, 1, 1, 1, 1]` (counts for 'I', 'love', 'natural', 'language', 'processing', 'is', 'fun')

### Why is BoW a foundational concept?

*   **Simplicity**: It's easy to understand and implement.
*   **Effectiveness**: Despite its simplicity, BoW can be surprisingly effective for many text classification, sentiment analysis, and topic modeling tasks, especially when combined with statistical models like Naive Bayes or SVMs.
*   **Feature Generation**: It provides a direct way to convert raw text into numerical features that can be fed into machine learning algorithms.

However, it loses information about word order and context, which can be a limitation for tasks requiring deeper linguistic understanding. Later, we will explore more advanced techniques that address these limitations.

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd # Import pandas for DataFrame display if not already imported

# Select a small subset of sentences from the DataFrame
sample_sentences = df['text'].iloc[0:5].tolist()

print("--- Original Sample Sentences ---")
for i, sentence in enumerate(sample_sentences):
    print(f"Document {i+1}: {sentence}")

# Initialize CountVectorizer
# For a simple demonstration, we'll keep default parameters. We can add stop_words, lowercase, etc. later.
vectorizer = CountVectorizer()

# Fit the vectorizer to the sample sentences and transform them into a BoW matrix
bow_matrix = vectorizer.fit_transform(sample_sentences)

# Get the vocabulary (feature names)
vocabulary = vectorizer.get_feature_names_out()

print("\n--- BoW Vocabulary ---")
print(vocabulary)

print("\n--- BoW Feature Matrix (Word Counts) ---")
# Convert the sparse matrix to a dense array and then to a DataFrame for better readability
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vocabulary)
display(bow_df)

print("Bag of Words representation demonstrated successfully.")
```

## Feature Extraction: TF-IDF

### Subtask:

## Feature Extraction: TF-IDF

While the Bag of Words (BoW) model provides a simple and effective way to represent text numerically, it has a limitation: it treats all words equally in terms of importance. A very common word like "the" or "is" would have a high count in many documents, but it doesn't necessarily carry more information than a less frequent, more domain-specific word.

This is where **TF-IDF (Term Frequency-Inverse Document Frequency)** comes in. TF-IDF is a numerical statistic that reflects how important a word is to a document in a collection or corpus. It is often used as a weighting factor in information retrieval and text mining.

### How TF-IDF Improves upon BoW
TF-IDF addresses the limitation of BoW by assigning a weight to each word, indicating its relevance. Words that appear frequently in a specific document but rarely in the rest of the corpus receive a higher TF-IDF score, making them more informative. Conversely, words common across many documents (like stopwords) get lower scores.

### Components of TF-IDF
TF-IDF is composed of two main parts:

1.  **Term Frequency (TF)**: This measures how frequently a term (word) appears in a document. Since documents may vary in length, TF is often normalized by dividing the raw count of a term by the total number of terms in the document. A high TF score means the word is prominent in that specific document.

    *   **Formula (simple form)**: `TF(t, d) = (Number of times term t appears in document d) / (Total number of terms in document d)`

2.  **Inverse Document Frequency (IDF)**: This measures how important a term is across the whole corpus. Rare terms typically have a high IDF score, while common terms (like "the", "a", "is") have a low IDF score. This helps to down-weight words that are too common.

    *   **Formula**: `IDF(t, D) = log_e(Total number of documents D / Number of documents d containing term t)`

### Calculating TF-IDF
The TF-IDF score is simply the product of the TF and IDF values:

`TF-IDF(t, d, D) = TF(t, d) * IDF(t, D)`

This results in a numerical representation where the values indicate the significance of a word within a document, relative to the entire dataset.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd # Ensure pandas is imported for DataFrame display

# Reuse the sample_sentences from the BoW demonstration
# sample_sentences = df['text'].iloc[0:5].tolist() # Assuming this list is already available from previous execution

print("--- Original Sample Sentences ---")
for i, sentence in enumerate(sample_sentences):
    print(f"Document {i+1}: {sentence}")

# Initialize TfidfVectorizer
# For a simple demonstration, we'll keep default parameters.
vectorizer_tfidf = TfidfVectorizer()

# Fit the vectorizer to the sample sentences and transform them into a TF-IDF matrix
tfidf_matrix = vectorizer_tfidf.fit_transform(sample_sentences)

# Get the vocabulary (feature names)
vocabulary_tfidf = vectorizer_tfidf.get_feature_names_out()

print("\n--- TF-IDF Vocabulary ---")
print(vocabulary_tfidf)

print("\n--- TF-IDF Feature Matrix ---")
# Convert the sparse matrix to a dense array and then to a DataFrame for better readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vocabulary_tfidf)
display(tfidf_df)

print("TF-IDF representation demonstrated successfully.")
```

## Word Embeddings: An Introduction

After converting text into numerical forms like Bag of Words or TF-IDF, a common challenge is that these representations are often very sparse (many zeros) and don't inherently capture semantic relationships between words. For example, BoW or TF-IDF don't know that "king" and "queen" are related, or that "running" and "jogging" have similar meanings.

### What are Word Embeddings?

**Word Embeddings** are dense vector representations of words that capture semantic and syntactic relationships. Each word is mapped to a vector of real numbers, typically in a low-dimensional space (e.g., 50 to 300 dimensions). The key idea is that words that appear in similar contexts tend to have similar meanings, and thus, their vector representations (embeddings) will be close to each other in the vector space.

**Key Characteristics:**

1.  **Dense Representation**: Unlike sparse BoW/TF-IDF vectors, word embeddings have values for most dimensions, making them more efficient and powerful.
2.  **Semantic Relationships**: The most remarkable property of word embeddings is their ability to capture semantic relationships. For example, in many embedding spaces, `vector('king') - vector('man') + vector('woman')` is approximately equal to `vector('queen')`.
3.  **Syntactic Relationships**: They can also capture grammatical relationships, such as verb tenses or pluralization.
4.  **Transfer Learning**: Pre-trained word embeddings (trained on massive text corpuses like Wikipedia or Google News) can be directly used in new NLP tasks, even with smaller datasets, which is a powerful form of transfer learning.

Popular word embedding models include Word2Vec, GloVe (Global Vectors for Word Representation), and FastText. These models learn these representations by analyzing patterns in large text datasets.

## Similarity Search

One of the powerful applications of numerical text representations, such as TF-IDF or word embeddings, is the ability to quantify the similarity between different pieces of text. This is fundamental to tasks like information retrieval, document clustering, and recommendation systems.

### What is Similarity Search?

Similarity search involves finding documents or text snippets that are most similar to a given query. In NLP, this often means transforming text into vectors (like BoW, TF-IDF, or word embeddings) and then calculating a distance or similarity metric between these vectors.

Common similarity metrics include:
*   **Cosine Similarity**: Measures the cosine of the angle between two non-zero vectors. A value of 1 means the vectors are identical, 0 means they are orthogonal (no similarity), and -1 means they are diametrically opposed.
*   **Euclidean Distance**: Measures the straight-line distance between two points in Euclidean space.

In this section, we will demonstrate basic keyword search and then dive into using cosine similarity with TF-IDF vectors to find documents that are semantically close to a given query.

### Basic Keyword Search (Baseline)

Before diving into more advanced similarity measures, let's consider a very simple form of similarity search: **keyword search**. This method directly looks for the presence of specific words or phrases within the text. While straightforward, it can be quite effective for initial filtering or when precise word matching is sufficient. It serves as a good baseline to understand the limitations that more advanced techniques like cosine similarity aim to overcome.

For this demonstration, we'll create a function that takes a query (a keyword or phrase) and returns sentences from our dataset that contain that query.

## Similarity Search: Keyword Search

```python
def keyword_search(query, dataframe, text_column='text'):
    """
    Performs a case-insensitive keyword search within a DataFrame's text column.

    Args:
        query (str): The keyword or phrase to search for.
        dataframe (pd.DataFrame): The DataFrame containing the text data.
        text_column (str): The name of the column containing the text data.

    Returns:
        list: A list of indices of documents that contain the query.
    """
    matching_documents = []
    print(f"Searching for '{query}' (case-insensitive):\n")

    for index, row in dataframe.iterrows():
        document_text = str(row[text_column]).lower()
        if query.lower() in document_text:
            matching_documents.append(index)
            print(f"  - Document Index {index}: {row[text_column]}\n")

    if not matching_documents:
        print("  No documents found matching the query.")
    print("\n---\n")
    return matching_documents

# Demonstrate the keyword search function
print("## Demonstrating Keyword Search\n")

# Example 1: Search for 'NLP'
keyword_search('NLP', df)

# Example 2: Search for 'machine learning'
keyword_search('machine learning', df)

# Example 3: Search for a less frequent word (e.g., 'dog')
keyword_search('dog', df)
```

## Similarity Search: Cosine Similarity

After understanding basic keyword search, let's delve into a more nuanced way of measuring text similarity using **Cosine Similarity**. This metric is fundamental in many NLP applications, including recommendation systems, document clustering, and information retrieval.

### What is Cosine Similarity?

Cosine similarity measures the cosine of the angle between two non-zero vectors in a multi-dimensional space. In the context of NLP, these vectors represent documents (or queries), and their dimensions correspond to words (as in BoW or TF-IDF).

*   **High Similarity (Angle close to 0°, Cosine close to 1)**: If two document vectors are pointing in roughly the same direction, they are considered highly similar. A cosine similarity of `1` means the vectors are identical (or one is a scalar multiple of the other), indicating maximum similarity.
*   **Low Similarity (Angle close to 90°, Cosine close to 0)**: If the vectors are orthogonal (perpendicular), they have no directional relationship, suggesting no similarity in content. A cosine similarity of `0` typically means the documents share no common terms (or terms with non-zero weights).
*   **Dissimilarity (Angle close to 180°, Cosine close to -1)**: While less common with frequency-based text vectors (which are usually non-negative), a cosine similarity of `-1` would imply that the vectors are diametrically opposed, indicating maximum dissimilarity.

**Why is it suitable for measuring document similarity in NLP?**

1.  **Independent of Document Length**: Cosine similarity is a measure of orientation, not magnitude. This means it's less affected by the length of the documents. Two documents discussing the same topic will have a small angle between their vectors, regardless of whether one document is much longer than the other.
2.  **Captures Semantic Orientation**: It focuses on the proportional presence of terms rather than their absolute counts. If two documents use similar words in similar proportions, their vectors will point in similar directions.
3.  **Effective with High-Dimensional Sparse Data**: It performs well with the high-dimensional, sparse vectors often generated by BoW or TF-IDF representations.

### Cosine Similarity vs. Distance Metrics

Unlike distance metrics (like Euclidean distance), which measure the absolute distance between two points, cosine similarity measures the *angle* between vectors. For text data, where document lengths can vary wildly, focusing on the angle (or direction) of the vectors is often more indicative of thematic similarity than the raw distance. A long document and a short document on the same topic might have a large Euclidean distance but a high cosine similarity, making it more appropriate for text analysis.

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd # Ensure pandas is imported for DataFrame operations

print("## Demonstrating Cosine Similarity with TF-IDF Vectors\n")

# 1. Re-initialize and fit TfidfVectorizer on the entire df['text'] column
#    Using the same vectorizer as before, but fitting on the full dataset now.
vectorizer_full_tfidf = TfidfVectorizer()
tfidf_matrix_full = vectorizer_full_tfidf.fit_transform(df['text'])

print(f"Shape of full TF-IDF matrix: {tfidf_matrix_full.shape}\n")

# 2. Define a sample query string
query_string = 'artificial intelligence and natural language processing'
print(f"Query: '{query_string}'\n")

# 3. Transform the query_string into a TF-IDF vector using the *same* fitted TfidfVectorizer instance
query_tfidf = vectorizer_full_tfidf.transform([query_string])
print(f"Shape of query TF-IDF vector: {query_tfidf.shape}\n")

# 4. Calculate the cosine similarity between query_tfidf and tfidf_matrix_full
cosine_similarities = cosine_similarity(query_tfidf, tfidf_matrix_full).flatten()

# 5. Get the indices of the documents sorted by their similarity scores in descending order
similarity_scores = pd.Series(cosine_similarities)
sorted_indices = similarity_scores.argsort()[::-1]

# 6. Print the query_string and then display the top N most similar documents
N = 3 # Display top 3 similar documents
print(f"Top {N} most similar documents to the query:\n")

for i, idx in enumerate(sorted_indices[:N]):
    print(f"Rank {i+1} (Score: {cosine_similarities[idx]:.4f}):")
    print(f"  Document Index {idx}: {df['text'].iloc[idx]}\n")

print("Cosine similarity demonstration completed successfully.")
```
